**A matrix is in row-echelon form if:**
* **All rows that contain only zeros are at the bottom of the matrix; correspondingly, all rows that contain at least one nonzero element are on top of row that contain only zeros**
* **Looking at nonzero rows only, the first nonzero number from the left(also called the *pivot* or the *leading coefficient*) is always strictly to the right of the pivot of the row above it**
* **The variables corresponding to the pivots in the row-echelon form are called basic variables and the other variables are *free variables***
* **In order to determine a particular solution in row-echelon form, we express the right-hand side of the equation system using the pivot columns, such that** $ \pmb {b} = \sum_{i=1}^{P} \lambda_i \pmb {p}_i $ **, where** $ \pmb {p}_i, i =1,...,P $ **are the pivot columns.**

* **Reduced Row Echelon Form**
    * **An equation system is in *reduced row-echelon form* if**
        * **It is in row-echelon form**
        * **Every pivot is 1**
        * **The pivot is the only nonzero entry in the column**

### 2.3.4 Algorithms for Solving a System of Linear Equations
* **Assumptions:**
    * **A solution exist.**
* **We can use the transformation**
$$ \pmb {Ax} = \pmb {b} \Leftrightarrow \pmb {A}^{T} \pmb {Ax} = \pmb {A}^{T} \pmb {Ax} = \pmb {A}^{T} \pmb {b} \Leftrightarrow \pmb {x} = (\pmb {A}^{T} \pmb {A})^{-1} \pmb {A}^{T} \pmb {b} $$
* **Applying the Moore-Penrose pseudo-inverse**
$$ (\pmb{A}^{T} \pmb {A})^{-1} \pmb {A}^{T} $$
    * **This is also the minimum norm least-squares solution**
    * **A disadvantage of this approach is that it requires many compuations fot the matrix-matrix product and computing the inverse of** $ \pmb {A}^{T} \pmb {A} $**
    * **For reasons of numerical precision it s generally not recommended to compute the inverse of pseudo-inverse**
* **Gaussian Elimination**
    * **The number of arithmetic operations scales cubically in the number of simultaneous equations**
* **In practice, systems of many linear equations are solved indirectly, by either stationary iterative methods, and the successive over-relaxation method.**
    * **Iterative methods: let** $ x_* $ **be a solution of** $ \pmb {Ax} = \pmb {b} $. **The key idea of the iterative methods is to set up an iteration of the form**

**for suitable** $ \pmb {C} $ **and** $ \pmb {d} $ **that reduces the residual error** $ \parallel \pmb {x}^{k+1}-\pmb {x}_{*} \parallel $ **in every iteration and converges to** $ \pmb {x}_{*} $

### 2.4 Vector Spaces

    1. **Closer of** $ \mathcal {G} $ **under** $ \otimes $
$$ \forall x,y \in \mathcal {G}: x \otimes y \in \mathcal {G} $$
    2. **Associativity**
$ \forall x,y,z \in \mathcal {G}: $
$$ (x \otimes y) \otimes z = x \otimes (y \otimes z) $$
    3. **Neutral element:**
$ \exists e \in \mathcal {G} \forall x \in \mathcal {G}: $
$$ x \otimes e = x\space\space and\space\space e \otimes x = x $$
    4. **Inverse element:**
$ \forall x \in \mathcal {G} \exists y \in \mathcal {G}: $
$$ x \otimes y = e \space \space and \space \space y \otimes x = e $$
**where** $ e $ **is the neutral element**

* **Abelian Group**
    **If additionally,** $ \forall x,y \in \mathcal {G}: x \otimes y = y \otimes x $ **,then** $ G=(\mathcal {G}, \otimes) $ **is an Abelian group**

### 2.4.2 Vector Spaces
* **We can think of the inner operations as a form of addition, and outer operaton as a form of scaling.**
* **Definition 2.9(Vector Space).**
    * **A real-valued *vector space*** $ V=(V,+,\cdot) $ **is a set** $ \mathcal {V} $ **with two operations**
$$ +: \mathcal {V} \times \mathcal {V} \to \mathcal {V} $$
$$ \cdot: \mathbb {R} \times \mathcal {V} \to \mathcal {V} $$
**where**<br>

1. $ ( \mathcal {V}, +) $ **is an Abelian group**
1. **Distributivity**
    1. $ \forall \lambda \in \mathbb {R}, \pmb {x,y} \in \mathcal {V} $:
$$ \lambda \cdot (\pmb {x} + \pmb {y}) = \lambda \cdot \pmb {x} + \lambda \cdot \pmb {y} $$
    1. $ \forall \lambda,\psi \in \mathbb {R}, \pmb {x} \in \mathcal {V}: $
$$ (\lambda + \psi)\cdot \pmb{x} = \lambda \cdot \pmb{x} + \psi \cdot \pmb {x} $$
1. **Associativity(Outer operation)**
$ \forall \lambda, \psi \in \mathbb {R}, \pmb {x} \in \mathcal {V}:$
$$ \lambda \cdot (\lambda \cdot \pmb {x}) = (\lambda \psi)\cdot \pmb {x} $$
1. **Neutral element with respect to the outer operation**
$\forall x \in \mathcal {V}: $
$$ 1 \cdot \pmb {x} = \pmb {x} $$
**The elements** $ \pmb {x} \in V $ **are called vectors. The neutral element of** $ (\mathcal {V}, +) $ **is the zero vector**  
* **The inner operation** $ + $ **is called *vector addition.* The elements** $ \lambda \in \mathbb {R} $ **are called *scalars* and the outer operation** $ \cdot $ **is a *multiplication by scalars**


### 2.4.3 Vector Subspaces
* **They are sets contained in the original vector space with the property that when we perform vector space operations on elements within this subpace, we will never leave it. In this sense, they are "closed".**
* **Let** $ \textit {V} = (\textit {V}, +, \dot) $ **be a vector space and** $ \mathcal {U} \subseteq \textit {V}, \mathcal {U} \neq 0 $. **Then** $ \textit {U} = (\mathcal {U},+,\cdot) $ **is called *vector subspace* of *V(or linear subspace)* if *U* is a vector space with the vector space operations** $ + $ **and** $ \cdot $ **restricted to** $ \mathcal {U} \times \mathcal {U} $ **and** $ \mathbb {R} \times \mathcal {U} $
* **To determine whether** $ (\mathcal {U}, +, \cdot) $ **is a subspace of** $ \textit{V} $ **we still need to show**
    * $ \mathcal {U} \neq 0 $ **in particular:** $ 0 \in \mathcal {U} $
    * **Closure of *U***:
        1. **With respect to the outer operation:** 
$$ \forall \lambda \in \mathbb {R} \forall x \in \mathcal {U}: \space\lambda \pmb {x} \in \mathcal{U} $$
        1. **With respect to the inner operation:**
$$ \forall \pmb {x,y} \in \mathcal {U}: \space \pmb{x} + \pmb{y} \in \mathcal {U} $$

* **Every subspace** $ \mathcal{U} \subseteq (\mathbb{R}^{n}, +, \cdot) $ **is the solution space of a homogeneous system of linear equaitions** $ \pmb{A}\pmb{x} = \pmb{0} \forall x \in \mathbb{R}^{n} $

## 2.5 Linear Independence
* **Linear combination**
    * **Consider a vector space** $ {V} $ **and a finite number of vectors** $ \pmb{x}_1,...,\pmb{x}_k \in \textit{V} $, **Then, every** $ \pmb {v} \in \textit {V} $ **of the form**
$$ \pmb{b} = \lambda_1\pmb{x}_1+\dots+\lambda_k\pmb{x}_k = \sum_{i=1}^{k}\lambda_ix_i\in\textit{V} $$
**with** $ \lambda_1,\dots,\lambda_k\in\mathbb{R} $ **is a linear combination of the vectors** $ \pmb{x}_1, \dots,\pmb{x}_k. $
    * **The 0-vector can always be written as the linear combination of** *k* **vectors, x_1,...,x_k because** $ \pmb{0} = sum_{i=1}^{k} 0\pmb{x_i} $ **is always true.**
    * **Non-trivial linear combinations of a set of vectors to represent** $ \pmb 0 $: **linear combinations of vectors** $ \pmb{x}_i, \dots,\pmb{x}_k $, **where not all coefficient** $ \lambda_i $ **are 0**
* **Linear Independence**
    * **Consider a vector space** $ \textit{V} $ **with** $ k \in \mathbb{N} $ **and** $ \pmb{x}_1,\dots,\pmb{x}_k \in \textit{V} $
        1. **If there is a non-trivial linear combination, such that** $ \pmb{0} = \sum_{i=1}^{k}\lambda_i\pmb{x}_i $ **with at least one** $ \lambda_i \neq 0 $, **the vectors** $ \pmb{x}_1,\dots,\pmb{x}_k $ **are linearly independent**
        1. **Intuitively, a set of linearly independent vectors consists of vectors that gave no redundancy, i.e., if we remove any of those vectors from the set, we will lose something**

**The following properties are useful to find out whether vectors are linearly independent**
* $ \textit{k} $ **vectors are either linearly dependent or linearly independent**
    * **If at least one of the vectors** $ \pmb{x}_1,\dots,\pmb{x}_k $ **is 0 then they are linearly dependet. The same holds if two vectors are identical**
    * **The vectors** $ {\pmb{x}_1,\dots,\pmb{x}_k:\pmb{x}_i\neq\pmb{0},\textit{i}=1,\dots,\textit{k}},k \geq 2 $, **are linearly dependent if and only if(at least) one of them is a linear combination of the others. In particular, if one vector is a multiple of another vector, i.e.,** $ \pmb{x}_i = \lambda\pmb{x}_j, \lambda \in\mathbb{R} $, **then set** $ \{\pmb{x}_1,\dots,\pmb{x}_k:\pmb{x}_i \neq \pmb{0}, \textit{i}=1,\dots,\textit{k}\} $ **is linearly dependent**
    * **A practical way of checking whether vectors** $ \pmb{x}_1,\dots,\pmb{x}_k\in\textit{V} $ **are linearly independent is to use Gaussian elimination: write all vectors as columns of a matrix** $ \pmb{A} $ **and perform Gaussian eleimination until the matrix is in row echelon form**
        * **The pivot columns indicate the vectors, which are linearly independent of the vectors on the left.**
        * **The non-pivot columns can be expressed as linear combinations of the pivot columns on their left**
* **Consider a vector space** $ V $ **with** $ \textit{k} $ **linearly independent vectors** $ \pmb{b}_1,\dots,\pmb{b}_k $ **and** m **linear combinations**
$$ \pmb{x}_1=\sum_{i=1}^{k}\lambda_{il}b_{i} $$
$$ \vdots $$
$$\pmb{x}_m=\sum_{i=1}^{k}\lambda_{im}b_{i} $$
**Defining** $ \pmb{B} = [\pmb{b}_1,\dots,\pmb{b}_k] $ **as the matrix whose columns are the linearly independent vectors** $ \pmb{b}_1,\dots,\pmb{b}_k $, **we can write**
$$ \pmb{x}_j=\pmb{B}\lambda_j, 
\pmb{\lambda}=
\begin{bmatrix}
\lambda_{ij}\\
\vdots\\
\lambda_{kj}
\end{bmatrix},
\textit{j}=1,\dots,\textit{m}
$$
* **To test whether** $ \pmb{x}_1,\dots,\pmb{x}_m $ **are linearly independent, we can use the general approach of testing when**
$ \sum_{j=1}^{m}\psi_j\pmb{x}_j=\pmb{0} $, **we obtain**
$$\sum_{j=1}^{m}\psi_j\pmb{x}_j=\sum_{j=1}^{m}\psi_j\pmb{B}\pmb{\lambda}_j = \pmb{B}\sum_{j=1}^{m}\psi_j\pmb{\lambda}_j $$
**This means that** $ \{\pmb{x}_1,\dots,\pmb{x}_m\} $ **are linearly independent if and only if column vectors** $\{\pmb{\lambda}_1,\dots,\pmb{\lambda}_m\} $ **are linearly independent**

### 2.6.1 Generating Set and Basis
* **Generating Set and Span**
    * **Consider a vector space** $ \textit{V} = (\textit{V},+,\cdot) $ **and set of vectors** $ \mathcal{A}=\{\pmb{x}_1,\dots,\pmb{x}_k\} $ **If every vector** $ \pmb{v} \in \mathcal{V} $ **can be expressed as a linear combination of** $ \pmb{x}_1,\dots,\pmb{x}_k, \mathcal{A} $ **is called a *generating set* of** $ \pmb {V} $. **The set of all linear combinations of vectors in** $\mathcal{A}$ **is called the *span* of** $ \mathcal{A} $. **If** $ \mathcal{A} $ **spans the vector space** $\mathcal{V} $, **we write** $ \mathcal{V}=span[\mathcal{A}] $ **or**
$ \mathcal{V}=span[\pmb{x}_1,\dots,\pmb{x}_k] $
* **Basis**
    * **Consider a vector space** $ \mathcal{V}=(\mathcal{V},+,\cdot) $ **and** $ \mathcal{A} \subset \mathcal{V} $. **A generating set of** $ \mathcal {A} $ **of** $\mathcal{V} $ **is called minimal if there exists no smaller set** $ \tilde{\mathcal{A}} \subsetneq \mathcal{A} \subseteq \mathcal{V} $ 
* **Every linearly independent generating set of *V* is minimal and is called a basis of V**
* **Let** $ \textit{V} = (\mathcal{V},+,\cdot) $ **be a vector space and** $ \mathcal{B} \subseteq \mathcal{V}, \mathcal{B} \neq 0 $. **Then, the following statements are equivalenet**
    * $ \mathcal{B} $ **is a basis of** $ \mathcal{V} $
    * $ \mathcal{B} $ **is a minimal generating set**
    * $ \mathcal{B} $ **is a maximal linearly independent set of vectors in**$\textit{V}$, **i.e., adding any other vector to this set will make it linearly dependent**
    * **Every vector** $\pmb{x}\in\textit{V}$ **is a linear combination of vectors from** $\textit{B}$,**and every linear combination is unique, i.e., with**
$$ \pmb{x}=\sum_{i=1}^{k}\lambda_i\pmb{b}_i=\sum_{i=1}^{k}\lambda_i\pmb{b}_i $$

* **Every vector space** $\textit{V} $ **possesses a basis** $\pmb{B}$. **The preceding examples show that there can be many bases of a vector space** $\textit{V}$, **i.e., there is no unique basis. However, all bases possess the same number of elements, the *basis vectors***
* **We consider finite-dimensional vector space** $\textit{V}$ **The *dimension* of** $\textit{V}$, **is the number of basis vectors of** $\textit{V}$, **and we write** $dim(\pmb{V})$. **If** $U\subseteq V$ **is a subspace of V, then** $ dim(U) \leq dim(V) $ **and** $dim(U)=dim(V)$ **if and only if** $ U = V$
    * **Intuively, the dimension of a vector space can be thought of as the number of independent directions in this vector space**
    * **The dimension of a vector space is not necessarily the number of elements in a vector**
        * **The dimension of a vector space is not necessarily the number of elements in a vector.**
* **A basis of subspace** $U=span[\pmb{x}_1,\dots,\pmb{x}_m]\subseteq \mathbb{R}^{n} $ **can be found by executing the following steps:**
    1. **Write the spanning vectors as column of a matrix** $\pmb{A}$
    1. **Determine the row-echelon form** $\pmb{A}$
    1. **The spanning vectors associated with the pivot columns are a basis of** $ U $

### 2.6.2 Rank
* **The number of linearly independent columns of a matrix** $ \pmb{A} \in \mathbb{R}^{m \times n} $ **equals the number of linearly independent rows and is called the rank of** $ A$ **and is denoted by** $\tau\kappa(\pmb{A})$
**The rank of a matrix has some important properties**
* $\tau\kappa(\pmb{A})=\tau\kappa(\pmb{A}^{T})$,**i.e., the column rank equals the row rank**
* **The columns of $\pmb{A} \in \mathbb{R}^{m \times n} $ span a subspace $ U \subseteq \mathbb{R}^{m} $ with $ dim(U) = \tau\kappa(\pmb{A})$ A basis of $U$ can be found by applying Gaussian elimination to $\pmb{A}$ to identify the pivot columns**
* **The rows of $\pmb{A} \in \mathbb{R}^{m\times n} $ span a subspace $ W \subseteq \mathbb{R}^n $ with $ dim(W)=\tau\kappa(\pmb{A}) $. A basis of W can be found by applying Gaussian elimination function**
* **For all $\pmb{A} \in \mathbb{R}^{n \times n} $ it holds that $\pmb{A}$ is regular(invertible) if and only if** $\tau\kappa(\pmb{A})=n$
* **For $\pmb{A} \in \mathbb{R}^{m \times n} $ and all $ b \in \mathbb{R}^{m}$ it holds that the linea equation system $ \pmb{Ax}=\pmb{b} $ can be solved if and only if $ \tau\kappa(\pmb{A})=\tau\kappa(\pmb{A}|\pmb{b})$ where $ \pmb{A}|\pmb{b} $ denotes the augmented system**
* **For $\pmb{A} \in \mathbb{R}^{m \times n}$ the subspace of solutions for $ \pmb{A}\pmb{x} = \pmb{0} $ possesses dimension $ n - \tau\kappa(\pmb{A})$. This subspace is called kernel or null space**
* **A matrix $ \pmb{A} \in \mathbb{R}^{m \times n} $ has full rank if is rank equals the largest possile rank for a matrix of the same dimensions. This means that the rank of a full-rank matrix is the lesser of the number of rows and columns, i.e., $ \tau\kappa(A)=min(m,n)$. A matrix is said to be *rank deficient* if it does not have full rank**

## 2.7 Linear Mappings
* **Consider two real vector spaces,** $V, W$. **A mapping $ \Phi: V \space \to \space W $ preserves the structure of the vector space if**

$$ \Phi(\pmb{x}+\pmb{y}) = \Phi(\pmb{x})+\Phi(\pmb{y})$$

$$ \Phi(\lambda \pmb{x}) = \lambda\Phi(\pmb{x}) $$

$$\forall x,y \in V\space and\space \lambda \in \mathbb{R} $$
* **Linear Mapping. For vector spaces $ V,W $ , a mapping $ \Phi: V \to W $ is called a linear mapping(or vector space homomorphism/linear transformation) if**
$$ \forall \pmb{x} \pmb{y} \in V \forall \lambda, \psi \in \mathbb{R}: \Phi(\lambda \pmb{x}+ \psi \pmb{x})=\lambda\Phi(\pmb{x})+\psi\Phi(\pmb{y}) $$
* **Injective, Surjective, Bijective: Consider a mapping $ \Phi: \textit{V} \to \textit{W} $, where $ \textit{V}, \textit{W} $ can be arbitrary sets. Then $ \Phi $ is called**
    * **Injective if:**
$$ \forall \pmb{x,y} \in \textit{V}: \Phi{\pmb{x}}=\Phi{\pmb{y}} \implies \pmb{x}=\pmb{y} $$
    * **Surjective if:**
$$ \Phi(\textit{V})=\textit{W} $$
    * **Biijective if it is injective and surjective**
    * **If $\Phi$ is surjective, then every element in $\textit{W}$ can be "reached" from $\textit{V}$ using** $\Phi$. 
    * **A bijective can be "undone", i.e., there exisits a mapping $ \Psi : \textit{W}\to \textit{V} $ so that $ \Psi \circ \Phi (\pmb{x})=\pmb{x} $. This mapping $ \Psi $ is then called the inverse of $\Phi $ can normally denoted by** $ \Phi^{-1}$
    * ***Isomorphism*:
    $$ \Phi: \textit{V}\to\textit{W}$$ linear and bijective**
    * ***Endomorphism:*
    $$ \Phi: \textit{V}\to\textit{V} $$ linear**
    * ***Automorphism:*
    $$ \Phi: \textit{V} \to \textit{V} $$ linear and bijective**
    * **We define $$ id_V: \textit{V} \to \textit{V}, \pmb{x} \mapsto \pmb{x} $$ as the *identity mapping or identity automorphism***
   
* ***Finite-dimensional vector space V and W are isomorphic if and only if*** $ dim(\textit{V}) = dim(\textit{W}) $
    * **The theorem states that there exists a linear, bijective mapping between two vector spaces of the same dimension. Vector spaces of the same dimension are kind of the same thing, as they can be transformed to each other without incurring any loss**
    * **Theorem allows us to treat $ \mathbb{R}^{m\times n} $ as $ \mathbb{R}^{mn} $**
    * **Consider vector spaces *V*, *W*, *X*.**
        * **For linear mappings: $ \Phi: \textit{V} \to \textit{W} $ and $ \Psi: \textit{W} \to \textit{X}, $ the mapping $\Psi \circ \Phi: \textit{V} \to \textit{W} $**
        * **If $ \Phi: V \to W, \Psi: V \to W $ are linear, then $ \Psi + \Phi $ and $ \lambda \in \mathbb{R} $, are linear**


### 2.7.1 Matrix Representation of Linear Mapping
* **Any *n*-dimensional vector space is isopmorphic to $ \mathbb{R}^n $.**
    * **Consider an *ordered basis* V**
$$ \textit{B} = (\pmb{b}_1,\dots,\pmb{b}_n) $$
    * **Consider a vector space $\textit{V}$ and ordered basis $\textit{B}=(\pmb{b}_1,\dots, \pmb{b}_n) $ of $ \textit{V}$. For any $\pmb{x}\in\textit{V}$ we obtain a unique representation(linear combination)
$$ \pmb{x}=\alpha_1\pmb{b}_1+\dots+\alpha_n\pmb{b}_n$$
of $\pmb{x}$ with respect $\textit{B}$. Then $\alpha_1,\dots,\alpha_n $ are the coordinates of $\pmb{x} $ with respect to $\textit{B} $, and the vector 
$$ \pmb{\alpha}=
\begin{bmatrix}
\alpha_1\\
\vdots\\
\alpha_n\\
\end{bmatrix}
\in 
\mathbb{R}^{n}
$$
is the coordinate vector/coordinate representation of x with respect to the ordered basis $\textit{B}$**

* **For an *n*-dimensional vector space $V$ and an ordered basis $B$ of $V$, the mapping $\Phi: \mathbb{R}^{n}\to V. \Phi(\pmb{e}_i)=\pmb{b}_i, \textit{i}=1,\dots,n$ is linear, where $(\pmb{e}_1,\dots,\pmb{e}_n)$ is the standard basis of $\mathbb{R}^n$**
* **Transformation Matrix**
    * **Consider a vector space $ V,W $ with corresponding(ordered) bases $ \textit{B}=(\pmb{b}_1,\dots,\pmb{b}_n) $ and $ C\space = (\pmb{C}_1,\dots,\pmb{C}_n)$. Moreover, we consider a linear mapping: $\Phi: \textit{V} \to \textit{W}$ For $ j\in\{1,\dots,n\},$
$$\Phi(\pmb{b}_j)=\alpha_{1j}\pmb{c}_1+\dots+\alpha_{mj}\pmb{c}_m=\sum_{i=1}^{m}\alpha_{ij}\pmb{c}_i $$ 
is the unique representation of $ \Phi (\pmb{b}_j) $ with respect to $C$. Then, we call the $m \times n$ matrix $ \pmb{A}_\Phi $, whose elements are given by
$$ A_\Phi(i,j) = \alpha_{ij}$$
the *transformation matrix* of $\Phi$ (with respect to the ordered bases $\textit{B}$ of $\textit{V}$ and $\textit{C}$ of $\textit{W}$)**
* **Consider (finite-dimensional) vector spaces $V,W$ with ordered bases $B,C$ and a linear mapping $\Phi: V\to W $ with transformation matrix $\pmb{A}_\Phi$ If $ \hat{\pmb{x}}$ is the coordinate vector of $\pmb{x}\in\textit{V} $ with repsect to $ B$ and $\hat{y}$ the coordinate vector $ \pmb{y}=\Phi(\pmb{x})$ with respect to $C$, then
$$ \hat{\pmb{y}}=\pmb{A}_\Phi\hat{\pmb{x}} $$
This means that the transformation matrix can be used to map coordiantes with respect to an ordered basis in $V$ to coordinates with respect to an ordered basis in $W$**



### 2.7.2 Basis Change
* **Consider two ordered bases:
$$ B=(\pmb{b}_1,\dots,\pmb{b}_n), \tilde{B}=(\tilde{\pmb{b}}_1,\dots,\tilde{\pmb{b}}_n) $$
of $V$ and two ordered bases
$$ C=(\pmb{c}_1,\dots,\pmb{c}_m), \tilde{C}=(\tilde{\pmb{c}}_1,\dots,\tilde{\pmb{c}}_m) $$
let $\pmb{A}_\Phi \in \mathbb{R}^{m \times n} $ be the transformation matrix of the linear mapping $ \Phi: V \to W $ with respect to the bases $B$ and $C$, and $ \tilde{\pmb{A}}_\Phi \in \mathbb{R}^{m \times n} $ is the corresponding transformation mapping with respect to $ \tilde{B} $ and $ \tilde{C} $.**
* **The corresponding transformation matrix $ \tilde{\pmb{A}}_\Phi$ with respect to the bases $ \tilde{B} $ and $ \tilde{C} $ is given as**
$$ \tilde{\pmb{A}}_\Phi=\pmb{T}^{-1}\pmb{A}_{\Phi}\pmb{S} $$
* ***Here, $ \pmb{S} \in \mathbb{R}^{n \times n} $ is the transformation matrix of $ id_v $ that maps coordiantes with respect to $ \tilde{B} $ onto coordinates with respect to $ B$, and $ \pmb{T} \in \mathbb{R}^{m \times n} $ is the transformation matrix of $ id_w$ that maps coordiantes with respect to $ \tilde{C} $ onto coordinates with respect to $C$***

* **Consider a homorphism $ \Phi: V\to W $ and ordered bases,$ B, \tilde{B}$ of $V$ and $C, \tilde{C}$ of $W$. When we perform a basis change from $ B$ to $\tilde{B}$ in $V$ and from $C$ to $\tilde{C}$ in $W$, we can determine the corresponding transformation matrix $ \pmb{A}_\Phi$ as follows**
    1. **We first find the matrix representation of the linear mapping $ \Psi_{B\tilde{B}}: V \to V $ that maps coordinates with respect to the new basis $\tilde{B}$ onto the (unique) coordinates with respect to the "old" basis $ B$ (in $ V$)**
    1. **Then, we use the transformation matrix $ \pmb{A}_\Phi $ of $ \Phi_{CB}: \textit{V} \to \textit{W} $ to map these coordinates onto the coordiantes with respect to $ C $ in $ W$.**
    1. **Finally, we use a linear mapping $ \Xi_{\tilde{C}\tilde{C}}: W \to W $ to map the coordinateds with respect to $C$ onto coordinates with respect to $\tilde{C}$**
    1. **We can express the linear mapping $ \Phi_{\tilde{C}\tilde{B}} $ as a composition of linear mappings that involve the "old" basis**
$$  \Phi_{\tilde{C}\tilde{B}}=\Xi_{\tilde{C}C} \circ \Phi_{CB} \circ \Psi_{B\tilde{B}}=\Xi_{C\tilde{C}}^{-1} \circ \Phi_{CB} \circ \Psi_{B\tilde{B}} $$
* **Equivalence: Two matrices $ \pmb{A}, \pmb{\tilde{A}} \in \mathbb{R}^{m \times n} $ are *equivalent* if there exist regular matrices $ \pmb{S} \in \mathbb{R}^{n \times n} $ and $ \pmb{T} \in \mathbb{R}^{m \times m}$, such that**
$$ \tilde{\pmb{A}} = \pmb{T}^{-1} \pmb{A} \pmb{S} $$
* **Similarity: Two matrices $ \pmb{A}, \pmb{\tilde{A}}, \in \mathbb{R}^{m \times n} $ are similar if there exists a regular matrix $ \pmb{S} \in \mathbb{R}^{n \times n} $ with $ \tilde{\pmb{A}}= \pmb{S}^{-1}\pmb{A}\pmb{S} $**
    * **Similar matrices are always equivalent. However, equivalent matrices are not necessarily similar**
* **For Linear mappings: $ \Phi: \textit{V} \to \textit{W} $ and $\Psi: \textit{W} \to \textit{V} $, the mapping $ \Psi \circ \Phi: \textit{V} \to \textit{x} $ is also linear. With transformation matrices $ \pmb{A_\Phi}$ and $ \pmb{A_\Psi} $ of the corresponding mappings, the overall transformation matrix is** $ \pmb{A_{\Phi\circ\Psi}} = \pmb{A_\Phi}\pmb{A_\Psi} $
* $\pmb{A_\Phi}$ **is the tranformation matrix of a linear mapping $\Phi_{CB}: \textit{V} \to \textit{W} $ with respect to the bases $B,C$.**
* $\tilde{\pmb{A}}_\Phi$ **is the transformation matrix of the linear mapping $\Phi_{\tilde{C}\tilde{B}}: \textit{V} \to \textit{W}$ with respect to the bases** $ \tilde{B}, \tilde{C}$ 
* $\tilde{S}$ **transformation matrix of a linear mapping $\Phi_{B\tilde{B}}: \textit{V} \to \textit{V} $ (automorphism) that represents $\tilde{B}$ in terms of $B$. Normally, $\Psi=id_v$ is the identity mapping in $V$.**
* $\pmb{T} $ **transformation matrix of a linear mapping $\Xi: \textit{W} \to \textit{W} $ (automorphism) that represents $ \tilde{C} $ in terms of $C$. Normally, $\Xi=id_{w}$ is the identity mapping in $W$.**
* **As a result,**
$$ \tilde{B}\to\tilde{C}=\tilde{B}\to\textit{B}\to\textit{C}\to\tilde{C}$$
$$\tilde{A}_\Phi=\pmb{T}^{-1}\pmb{A}_\Phi\pmb{S}$$

### 2.7.3 Image and Kernel
* **For $\Phi: \textit{V} \to \textit{W}$, we define the *kernel/null space* 
$$ ker(\Phi):=\Phi^{-1}(\pmb{0}_w)=\{\pmb{v}\in V: \Phi(\pmb{v})=\pmb{0}_w\}$$
and the image/range
$$ Im(\Phi):=\Phi(\textit{V})=\{\pmb{w}\in W|\exists\pmb{v}\in V: \Phi(\pmb{v})=\pmb{w}$$
We also call $V$ and $W$ the *domain* and *codomain* of $\Phi$, respectively**
* **Intuitively, the kernel is the set of vectors $\pmb{v} \in V$ that maps onto the neutral element $\pmb{0}_w\in W$. The image is the set of vectors $ \pmb{w}\in W $ that can be "reached" by $\Phi$ from any vector in $V$**

* **Consider a linear mapping $\Phi: V\to W$, where $V, W$ are vector spaces**
    * **It always holds that $\Phi(\pmb{0}_v) = \pmb{0}_w$ and, therefore, $\pmb{0}_v\in ker(\Phi)$. In particular, the null space is never empty**
    * $Im(\Phi)\subseteq \textit{W} $ **is a subpsace of $W$, and $ker(\Phi) \subseteq V$ is a subspace of $V$**
    * $ \Phi$ **is injective(one to one) if and only if** $ker(\Phi)=\{\pmb{0}\}$
* **Null space and Column space: Let us consider $\pmb{A} \in \mathbb{R}^{m \times n} $ and linear mapping $\Phi: \mathbb{R}^{n} \to \mathbb{R}^{m}, \pmb \mapsto \pmb{Ax}$**
    * **For $\pmb{A}=[a_1,\dots,a_n]$, where $a_i$ are the columns of $\pmb{A}$, we obtain:
$$ Im(\Phi)=\{\pmb{Ax}:\pmb{x}\in\mathbb{R}^{n}\}=\{\sum_{i=1}^{n}x_i\pmb{a}_i:x_1,\dots,x_n\in\mathbb{R}\}$$
i.e., The image is the space of the columns of $\pmb{A}$, also called the *column space*. Therefore, the column space(image) is a subspace of $\mathbb{R}^{m}$, where $m$ is the "height" of the matrix**
$$ \tau\kappa(\pmb{A})=dim(Im(\Phi))$$
* **The kernel\null space $ker(\Phi) $ is the general solution to the homogeneous system of linear equations $\pmb{Ax}=\pmb{0}$ and captures all possible linear combinations of the elements in $\mathbb{R}^{n} $ that produce $\pmb{0}\in\mathbb{R}^{m}$**
* **The kernel is a subspaace of $\mathbb{R}^{n}$, where $n$ is the "width" of the matrix**
* **The kernel focuses on the relationship among the columns, and we can use it to determine whether/how we can express a column as a linear combination of other columns**
* **(Rank-Nullity Theorem). *For vector spaces $V,W$ and a linear mapping $ \Phi: \textit{V}\to\textit{W} $ it holds that*
$$ dim(ker(\Phi))+dim(Im(\Phi))=dim(\textit{V})$$**
     * **The rank-nullity theorem is also referred to as the fundamental theorem of linear mappings. The followings are direct consequences of the theorem**
    * **If $dim(Im(\Phi)) < dim(\textit{V})$, then $ker(\Phi)$ is non-trivial, i.e., the kernel contains more than $\pmb{0}_V$ and $dim(ker(\Phi))\geq 1$**
    * **If $ \pmb{A}_\Phi $ is the transformation matrix of $\Phi$ with respect to an ordered basis and $\dim(\Im(\Phi)) <dim(V)$, then the system of linear equations $\pmb{A}_\Phi\pmb{x}=\pmb{0} $ has infinitely many solutions**
    * **If $dim(V)=dim(W)$, then the following three-wat equivalence holds:**
         * $\Phi $ **is injective**
         * $\Phi$ **is surjective**
         * $\Phi$ **is bijective**
         * **since $ Im(\Phi)\subseteq W$**

## 2.8 Affine Spaces
### 2.8.1 Affine Subspaces
* **Let $ V$ be a vector space, $x_0 \in V $ and $U \subseteq V$ a subspace.Then the subset
$$ L=\pmb{x}_0 + U :=\{\pmb{x}_0+\pmb{u}:\pmb{u}\in U\}
    =\{\pmb{v}\in V|\exists\pmb{u}\in U: \pmb{v}=\pmb{x}_0+\pmb{u}\}\subseteq V$$
is called *affine subspace* or *linear manifold* of V. U is caled *direction* or *direction space*, and $\pmb{x}_0 $ is called *support point***
* **Note that the definition of an affine subspace excludes $ \pmb{0} \nsubseteq U$. Therefore, an affine subspace is not a (linear) subspace(vector subspace) of $V$ for $\pmb{x}_0\nsubseteq U$**
* **Consider two affine subspace $L=\pmb{x}_0 + U$ and $ \tilde{L}=\tilde{\pmb{x}}_0+\tilde{U}$ of a vector space V. Then, $L \subseteq \tilde{L} $ if and only if $U \subseteq \tilde{U} $ and $\pmb{x}-\pmb{\tilde{x}}\in \tilde{U} $** 
* **Affine subspaces are often described by *parameters*: consider a *k*-dimensional affine space $L =\pmb{x}_0 + U $ of $V$. If ($\pmb{b}_1,\dots,\pmb{b}_k$) is an ordered basis of $U$, then every elment $\pmb{x}\in L $ can be uniquely described as
$$ \pmb{x}=\pmb{x}_0+\lambda_1\pmb{b}_1+\dots+\lambda_k\pmb{b}_k$$
where $\lambda_1,\dots,\lambda_k\in \mathbb{R}$. This representation is called *parametric equation* of L with directional vectors $\pmb{b}_1,\dots,\pmb{b}_k$ and parameters $\lambda_1,\dots,\lambda_k$.**


* **For $A\in\mathbb{R}^{m \times n} $ and $\pmb{x}\in \mathbb{R}^{m}$, the solution of the system of linear equations $\pmb{A\lambda}=\pmb{x}$ is either the empty set ot an affine subspace of $\mathbb{R}^{n}$ of dimension $n-\tau\kappa(\pmb{A})$.**
* **In $\mathbb{R}^{n}$, every *k*-dimensional affine subspace is the solution of an inhomogeneous system of linear equations $\pmb{Ax}=\pmb{b}$, where $\pmb{A}\in\mathbb{R}^{m\times n}, b \in \mathbb{R}^{m} $ and $\tau\kappa(\pmb{A})=n-k$.**

### 2.8.2 Affine Mappings
* **Affine Mapping: For two vectors $V,W$, a linear mapping $\Phi: V\to W$, and $\pmb{a}\in W$, the mapping
$$ \phi: V\to W$$
$$ \pmb{x}\mapsto\alpha+\Phi(x)$$
is an affine mapping from $V$ to $W$. The vector $a$ is called the *translation vector* of $\phi$.**
* **Every affine mapping $\phi: V \to W$ is also the compositipon of a linear mapping $ \Phi: V\to W$ and a translation $\tau: W\to W $ in $W$, such that $\phi = \tau \circ \Phi$. The mapping $\Phi$ and $\tau$ are uniquely determined**
* **The composition $\phi^{'} \circ \phi$ of affine mappings $\phi:V\to W, \phi^{'}:W\to X $ is affine**
* **Affine mappings keep the geometric structure invariant. Theu also serve the dimension parallelism**
