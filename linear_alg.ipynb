**A matrix is in row-echelon form if:**
* **All rows that contain only zeros are at the bottom of the matrix; correspondingly, all rows that contain at least one nonzero element are on top of row that contain only zeros**
* **Looking at nonzero rows only, the first nonzero number from the left(also called the *pivot* or the *leading coefficient*) is always strictly to the right of the pivot of the row above it**
* **The variables corresponding to the pivots in the row-echelon form are called basic variables and the other variables are *free variables***
* **In order to determine a particular solution in row-echelon form, we express the right-hand side of the equation system using the pivot columns, such that** $ \pmb {b} = \sum_{i=1}^{P} \lambda_i \pmb {p}_i $ **, where** $ \pmb {p}_i, i =1,...,P $ **are the pivot columns.**

* **Reduced Row Echelon Form**
    * **An equation system is in *reduced row-echelon form* if**
        * **It is in row-echelon form**
        * **Every pivot is 1**
        * **The pivot is the only nonzero entry in the column**

### 2.3.4 Algorithms for Solving a System of Linear Equations
* **Assumptions:**
    * **A solution exist.**
* **We can use the transformation**
$$ \pmb {Ax} = \pmb {b} \Leftrightarrow \pmb {A}^{T} \pmb {Ax} = \pmb {A}^{T} \pmb {Ax} = \pmb {A}^{T} \pmb {b} \Leftrightarrow \pmb {x} = (\pmb {A}^{T} \pmb {A})^{-1} \pmb {A}^{T} \pmb {b} $$
* **Applying the Moore-Penrose pseudo-inverse**
$$ (\pmb{A}^{T} \pmb {A})^{-1} \pmb {A}^{T} $$
    * **This is also the minimum norm least-squares solution**
    * **A disadvantage of this approach is that it requires many compuations fot the matrix-matrix product and computing the inverse of** $ \pmb {A}^{T} \pmb {A} $**
    * **For reasons of numerical precision it s generally not recommended to compute the inverse of pseudo-inverse**
* **Gaussian Elimination**
    * **The number of arithmetic operations scales cubically in the number of simultaneous equations**
* **In practice, systems of many linear equations are solved indirectly, by either stationary iterative methods, and the successive over-relaxation method.**
    * **Iterative methods: let** $ x_* $ **be a solution of** $ \pmb {Ax} = \pmb {b} $. **The key idea of the iterative methods is to set up an iteration of the form**

**for suitable** $ \pmb {C} $ **and** $ \pmb {d} $ **that reduces the residual error** $ \parallel \pmb {x}^{k+1}-\pmb {x}_{*} \parallel $ **in every iteration and converges to** $ \pmb {x}_{*} $

### 2.4 Vector Spaces

    1. **Closer of** $ \mathcal {G} $ **under** $ \otimes $
$$ \forall x,y \in \mathcal {G}: x \otimes y \in \mathcal {G} $$
    2. **Associativity**
$ \forall x,y,z \in \mathcal {G}: $
$$ (x \otimes y) \otimes z = x \otimes (y \otimes z) $$
    3. **Neutral element:**
$ \exists e \in \mathcal {G} \forall x \in \mathcal {G}: $
$$ x \otimes e = x\space\space and\space\space e \otimes x = x $$
    4. **Inverse element:**
$ \forall x \in \mathcal {G} \exists y \in \mathcal {G}: $
$$ x \otimes y = e \space \space and \space \space y \otimes x = e $$
**where** $ e $ **is the neutral element**

* **Abelian Group**
    **If additionally,** $ \forall x,y \in \mathcal {G}: x \otimes y = y \otimes x $ **,then** $ G=(\mathcal {G}, \otimes) $ **is an Abelian group**

### 2.4.2 Vector Spaces
* **We can think of the inner operations as a form of addition, and outer operaton as a form of scaling.**
* **Definition 2.9(Vector Space).**
    * **A real-valued *vector space*** $ V=(V,+,\cdot) $ **is a set** $ \mathcal {V} $ **with two operations**
$$ +: \mathcal {V} \times \mathcal {V} \to \mathcal {V} $$
$$ \cdot: \mathbb {R} \times \mathcal {V} \to \mathcal {V} $$
**where**<br>

1. $ ( \mathcal {V}, +) $ **is an Abelian group**
1. **Distributivity**
    1. $ \forall \lambda \in \mathbb {R}, \pmb {x,y} \in \mathcal {V} $:
$$ \lambda \cdot (\pmb {x} + \pmb {y}) = \lambda \cdot \pmb {x} + \lambda \cdot \pmb {y} $$
    1. $ \forall \lambda,\psi \in \mathbb {R}, \pmb {x} \in \mathcal {V}: $
$$ (\lambda + \psi)\cdot \pmb{x} = \lambda \cdot \pmb{x} + \psi \cdot \pmb {x} $$
1. **Associativity(Outer operation)**
$ \forall \lambda, \psi \in \mathbb {R}, \pmb {x} \in \mathcal {V}:$
$$ \lambda \cdot (\lambda \cdot \pmb {x}) = (\lambda \psi)\cdot \pmb {x} $$
1. **Neutral element with respect to the outer operation**
$\forall x \in \mathcal {V}: $
$$ 1 \cdot \pmb {x} = \pmb {x} $$
**The elements** $ \pmb {x} \in V $ **are called vectors. The neutral element of** $ (\mathcal {V}, +) $ **is the zero vector**  
* **The inner operation** $ + $ **is called *vector addition.* The elements** $ \lambda \in \mathbb {R} $ **are called *scalars* and the outer operation** $ \cdot $ **is a *multiplication by scalars**


### 2.4.3 Vector Subspaces
* **They are sets contained in the original vector space with the property that when we perform vector space operations on elements within this subpace, we will never leave it. In this sense, they are "closed".**
* **Let** $ \textit {V} = (\textit {V}, +, \dot) $ **be a vector space and** $ \mathcal {U} \subseteq \textit {V}, \mathcal {U} \neq 0 $. **Then** $ \textit {U} = (\mathcal {U},+,\cdot) $ **is called *vector subspace* of *V(or linear subspace)* if *U* is a vector space with the vector space operations** $ + $ **and** $ \cdot $ **restricted to** $ \mathcal {U} \times \mathcal {U} $ **and** $ \mathbb {R} \times \mathcal {U} $
* **To determine whether** $ (\mathcal {U}, +, \cdot) $ **is a subspace of** $ \textit{V} $ **we still need to show**
    * $ \mathcal {U} \neq 0 $ **in particular:** $ 0 \in \mathcal {U} $
    * **Closure of *U***:
        1. **With respect to the outer operation:** 
$$ \forall \lambda \in \mathbb {R} \forall x \in \mathcal {U}: \space\lambda \pmb {x} \in \mathcal{U} $$
        1. **With respect to the inner operation:**
$$ \forall \pmb {x,y} \in \mathcal {U}: \space \pmb{x} + \pmb{y} \in \mathcal {U} $$

* **Every subspace** $ \mathcal{U} \subseteq (\mathbb{R}^{n}, +, \cdot) $ **is the solution space of a homogeneous system of linear equaitions** $ \pmb{A}\pmb{x} = \pmb{0} \forall x \in \mathbb{R}^{n} $

## 2.5 Linear Independence
* **Linear combination**
    * **Consider a vector space** $ {V} $ **and a finite number of vectors** $ \pmb{x}_1,...,\pmb{x}_k \in \textit{V} $, **Then, every** $ \pmb {v} \in \textit {V} $ **of the form**
$$ \pmb{b} = \lambda_1\pmb{x}_1+\dots+\lambda_k\pmb{x}_k = \sum_{i=1}^{k}\lambda_ix_i\in\textit{V} $$
**with** $ \lambda_1,\dots,\lambda_k\in\mathbb{R} $ **is a linear combination of the vectors** $ \pmb{x}_1, \dots,\pmb{x}_k. $
    * **The 0-vector can always be written as the linear combination of** *k* **vectors, x_1,...,x_k because** $ \pmb{0} = sum_{i=1}^{k} 0\pmb{x_i} $ **is always true.**
    * **Non-trivial linear combinations of a set of vectors to represent** $ \pmb 0 $: **linear combinations of vectors** $ \pmb{x}_i, \dots,\pmb{x}_k $, **where not all coefficient** $ \lambda_i $ **are 0**
* **Linear Independence**
    * **Consider a vector space** $ \textit{V} $ **with** $ k \in \mathbb{N} $ **and** $ \pmb{x}_1,\dots,\pmb{x}_k \in \textit{V} $
        1. **If there is a non-trivial linear combination, such that** $ \pmb{0} = \sum_{i=1}^{k}\lambda_i\pmb{x}_i $ **with at least one** $ \lambda_i \neq 0 $, **the vectors** $ \pmb{x}_1,\dots,\pmb{x}_k $ **are linearly independent**
        1. **Intuitively, a set of linearly independent vectors consists of vectors that gave no redundancy, i.e., if we remove any of those vectors from the set, we will lose something**

**The following properties are useful to find out whether vectors are linearly independent**
* $ \textit{k} $ **vectors are either linearly dependent or linearly independent**
    * **If at least one of the vectors** $ \pmb{x}_1,\dots,\pmb{x}_k $ **is 0 then they are linearly dependet. The same holds if two vectors are identical**
    * **The vectors** $ {\pmb{x}_1,\dots,\pmb{x}_k:\pmb{x}_i\neq\pmb{0},\textit{i}=1,\dots,\textit{k}},k \geq 2 $, **are linearly dependent if and only if(at least) one of them is a linear combination of the others. In particular, if one vector is a multiple of another vector, i.e.,** $ \pmb{x}_i = \lambda\pmb{x}_j, \lambda \in\mathbb{R} $, **then set** $ \{\pmb{x}_1,\dots,\pmb{x}_k:\pmb{x}_i \neq \pmb{0}, \textit{i}=1,\dots,\textit{k}\} $ **is linearly dependent**
    * **A practical way of checking whether vectors** $ \pmb{x}_1,\dots,\pmb{x}_k\in\textit{V} $ **are linearly independent is to use Gaussian elimination: write all vectors as columns of a matrix** $ \pmb{A} $ **and perform Gaussian eleimination until the matrix is in row echelon form**
        * **The pivot columns indicate the vectors, which are linearly independent of the vectors on the left.**
        * **The non-pivot columns can be expressed as linear combinations of the pivot columns on their left**
* **Consider a vector space** $ V $ **with** $ \textit{k} $ **linearly independent vectors** $ \pmb{b}_1,\dots,\pmb{b}_k $ **and** m **linear combinations**
$$ \pmb{x}_1=\sum_{i=1}^{k}\lambda_{il}b_{i} $$
$$ \vdots $$
$$\pmb{x}_m=\sum_{i=1}^{k}\lambda_{im}b_{i} $$
**Defining** $ \pmb{B} = [\pmb{b}_1,\dots,\pmb{b}_k] $ **as the matrix whose columns are the linearly independent vectors** $ \pmb{b}_1,\dots,\pmb{b}_k $, **we can write**
$$ \pmb{x}_j=\pmb{B}\lambda_j, 
\pmb{\lambda}=
\begin{bmatrix}
\lambda_{ij}\\
\vdots\\
\lambda_{kj}
\end{bmatrix},
\textit{j}=1,\dots,\textit{m}
$$
* **To test whether** $ \pmb{x}_1,\dots,\pmb{x}_m $ **are linearly independent, we can use the general approach of testing when**
$ \sum_{j=1}^{m}\psi_j\pmb{x}_j=\pmb{0} $, **we obtain**
$$\sum_{j=1}^{m}\psi_j\pmb{x}_j=\sum_{j=1}^{m}\psi_j\pmb{B}\pmb{\lambda}_j = \pmb{B}\sum_{j=1}^{m}\psi_j\pmb{\lambda}_j $$
**This means that** $ \{\pmb{x}_1,\dots,\pmb{x}_m\} $ **are linearly independent if and only if column vectors** $\{\pmb{\lambda}_1,\dots,\pmb{\lambda}_m\} $ **are linearly independent**

### 2.6.1 Generating Set and Basis
* **Generating Set and Span**
    * **Consider a vector space** $ \textit{V} = (\textit{V},+,\cdot) $ **and set of vectors** $ \mathcal{A}=\{\pmb{x}_1,\dots,\pmb{x}_k\} $ **If every vector** $ \pmb{v} \in \mathcal{V} $ **can be expressed as a linear combination of** $ \pmb{x}_1,\dots,\pmb{x}_k, \mathcal{A} $ **is called a *generating set* of** $ \pmb {V} $. **The set of all linear combinations of vectors in** $\mathcal{A}$ **is called the *span* of** $ \mathcal{A} $. **If** $ \mathcal{A} $ **spans the vector space** $\mathcal{V} $, **we write** $ \mathcal{V}=span[\mathcal{A}] $ **or**
$ \mathcal{V}=span[\pmb{x}_1,\dots,\pmb{x}_k] $
