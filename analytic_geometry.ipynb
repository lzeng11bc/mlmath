# Analytic Geometry

## 3.1 Norms
* **Norm: a norm on a vector space $ V $ is a function:
$$ \| \cdot \|: V \to \mathbb{R}, $$
$$ \pmb{x} \mapsto \| \pmb{x} \| $$
which assignes each vector $ x $ its *length* $ \| \pmb{x} \| \in \mathbb{R} $, such that for all $ \lambda \in \mathbb{R} $ and $ \pmb{x,y} \in V$ the following hold:**
    * ***Absolutely homogeneous:***
$$ \| \lambda \pmb{x} \| = \mid \lambda \mid \| \pmb{x} \| $$
    * ***Triangle inequality***
$$ \| \pmb{x} + \pmb{y} \| \leq \|\pmb{x}\| + \|\pmb{y}\| $$
    * ***Positive definite***
$$ \|\pmb{x}\| \geq 0\space and\space \|\pmb{x}\| = 0 \Leftrightarrow \pmb{x} = 0 $$

* **The *Manhattan norm* on $ \mathbb{R}^{n} $ is defined for $ \pmb{x} \in \mathbb{R}^{n} $ as**
$$ \|\pmb{x} \|_1 := \sum_{i = 1}^{n} \mid x_i \mid $$,
**where $ \mid \cdot \mid $ is the absolute value.**

**Shows all vectors $ \pmb{x} \in \mathbb{R}^2 $ with $ \mid \pmb{x} \mid = 1$.**
    * **The Manhattan norm is also called $\textit{l}_1$ norm**
* **The *Euclidean norm* of $ x \in \mathbb{R}^{n} $ is defined as**
$$ \|\pmb{x}\|_2 := \sqrt{\sum_{i = 1}^{n}x_i^2}=\sqrt{\pmb{x}^T\pmb{x}} $$ 
**and computes the *Eculidean distance* of $\pmb{x}$ from the origin**

**shows all vectors $ \pmb{x} \in \mathbb{R}^2 $  with $ \| \pmb{x} \|_2 = 1 $. The Euclidean norm is also called $ \mathit{l}_2 $ norm**

## 3.2 Inner Product
### 3.2.1 Dot Product
* **Scalar product/dot product in  $ \mathbb{R}^{n} $, which is given by**
$$ \pmb{x}^{T}\pmb{y} = \sum_{i = 1}^{n} x_i y_i $$

### 3.2.2 General Inner Products
* ***Bilinear mapping* $ \Omega $ is a mapping with two arguments, and it is linear in each argument, i.e., when we look at a vector space $ V $ then it holds that for all $ \pmb{x,y,z} \in V, \lambda, \psi \in \mathbb{R} $ that**
$$ \Omega (\lambda \pmb{x} + \psi \pmb{y, z}) = \lambda \Omega (\pmb{x,z}) + \psi \Omega (\pmb{y,z}) $$
$$ \Omega (\pmb{x}, \lambda\pmb{y} + \psi \pmb{z}) = \lambda\Omega(\pmb{x,y}) + \psi \Omega(\pmb{x,z}) $$
* **Let $ V $ be a vector space and $ \Omega: V \times V \to \mathbb{R} $ be a bilinear mapping that takes two vectors and maps them onto real number.**
    * $ \Omega $ **is called *symmetric* if $ \Omega(\pmb{x,y}) = \Omega(\pmb{y,x}) $ for all $ \pmb{x,y} \in V $**
    * $ \Omega $ **is called *positive definite* if**
$$ \forall \pmb{x} \in V \backslash \{0\} {x}: \Omega (\pmb{x,x}) > 0,\space \space \Omega(\pmb{0,0}) = 0$$
* **Let $V$ be a vector space and $\Omega: V \times V \to \mathbb{R} $ be a bilinear mapping that takes two vectors and maps them onto a real number. Then**
    * **A positive definite, symmetric bilinear mapping $ \Omega: V \times V \to \mathbb{R} $ is called an *inner* product on $V$. We tyically write $\langle \pmb{x,y} \rangle $ instead .**
    * **The pair $ (V, \langle \cdot , \cdot \rangle) $ is called an *inner product space* or (real) *vector space* with *inner product*. We call $ (V, \langle \cdot, \cdot \rangle) $ a *Euclidean vector space***
    * **Consider $ V = \mathbb{R}^{2}$
    $$ \langle \pmb{x,y} \rangle := x_1y_1 -(x_1y_2 + x_2y_1) + 2 x_2 y_2 $$
then $ \langle \cdot, \cdot \rangle $ is an inner product but different from the dot product**
* **Consider an *n*-dimensional vector space $ V $ with an inner product $\langle \cdot, \cdot \rangle : V \times V \to \mathbb{R} $ and an ordered basis $ B = (\pmb{b}_1, \dots, \pmb{b}_n $ of $ V$. Due to the bilinearity of the inner product, it holds for all $ \pmb{x,y} \in V $ that
$$ \langle\pmb{x,y} \rangle = \langle \sum_{i=1}^{n} \psi_i \pmb{b}_i, \sum_{j = 1}^{n} \lambda_i \pmb{b}_j \rangle = \sum_{i = 1}^{n} \sum_{j=1}^{n} \psi_i \langle\pmb{b}_i, \pmb{b}_j \rangle \lambda_j = \hat{\pmb{x}}^{T} \pmb{A} \hat{\pmb{y}} $$
where $ A_ij := \langle \pmb{b}_i, \pmb{b}_j \rangle $ and $ \hat{\pmb{x}}, \hat{\pmb{y}} $ are the coordinates of $ \pmb{x} $ and $ \pmb{y} $ with respect to the basis $ B$.**
    * **Thid implies that the inner product $ \langle \cdot, \cdot, \rangle $ is uniquely determined through $ \pmb{A} $. The symmetry of the inner product also means that $ \pmb{A} $ is symmetric**
    * **The positive definiteness of the inner product implies that**
$$ \forall \pmb{x} \in V \backslash\{0\}: \pmb{x}^{T}\pmb{A} \pmb{x} > 0 $$
    * **A symmetric matrix $ \pmb{A} \in \mathbb{R}^{n \times n} $ that satisfies the above equation is called *symmetrix, positive definie*, or just *positive definite*. If only $\geq$ holds, then $\pmb{A} $ is called *symmetrix, positive semidefinite***
    * **If $ \pmb{A} \in \mathbb{R}^{n \times n} $ is symmetric, positive definite, then
$$ \langle \pmb{x,y} \rangle = \hat{\pmb{x}}^{T} \pmb{A} \hat{\pmb{y}} $$
defines an inner product with respect to an ordered basis $ B$, where $\hat{x} $ and $ \hat{y} $ are the coordicate representation of $\pmb{x,y} \in V $ with respect to $B$**
* **For a real-values, finite-dimensional vector space $V$ and an ordered basis $B$ of $V$, it holds that $ \langle \cdot, \cdot \rangle: V \times V \to \mathbb{R} $ is an inner product if and only if there exists a symmetric, definite matrix $ \pmb{A} \in \mathbb{R}^{n \times n} $ with**
$$ \langle \pmb{x,y} \rangle = \hat{\pmb{x}}^{T} \pmb{A} \hat{\pmb{y}} $$
* **The following properties hold if $\pmb{A} \in \mathbb{R}^{n \times n} $ is symmetric and positive definite:**
    * **The null space(kernel) of $ \pmb{A} $ consists only of $ \pmb{0} $ because $ \pmb{x}^{T} \pmb{A} \pmb{x} > 0$ for all $ \pmb{x} \neq \pmb{0} $. This implies that $ \pmb{A} \pmb{x} \neq \pmb{0} $ id $ \pmb{x} \neq 0 $**
    * **The diagonal elements $ a_{ii} $ of $ \pmb{A} $ are positive because $ a_{ii} = \pmb{e}^{T} \pmb{A} \pmb{e}_i > 0$, where $\pmb{e}_i$ is the ith vecotr of the standard basis in $\mathbb{R}^ n$**

## 3.3 Length and Distances
* **Cauchy-Schwarz Inequality. For an inner product vector space $ ( V, \langle \cdot , \cdot \rangle ),$ the induced norm $ \| \cdot \| $ satisfies the *Cauchy-Shwarz Inequality***
$$ \langle \pmb{x,y} \rangle \leq  \|\pmb{x}\| \|\pmb{y}\| $$
* **Distnace and Metric. Consider an inner product space $ (V, \langle \dot, \dot \rangle ) $. Then 
$$ d(\pmb{x,y} ) = \|\pmb{x} - \pmb{y} \| = \sqrt{\langle \pmb{x} -\pmb{y}, \pmb{x} -\pmb{y} \rangle} $$
is called the *distacne* between $ \pmb{x} $ and $ \pmb{y} $ for $ \pmb{x, y} \in V $. If we use the dot product as the inner product, then the distance is called *Euclidean distance***
* **The mapping
$$ d: V \times V \to \mathbb{R} \\
 (\pmb{x,y}) \mapsto d(\pmb{x,y})$$
is called a *metric***
* **The distance between vectors does not require an inner product: a norm is sufficient**
* **The metric $ d $ satisfies the following**
    * $ d $ **is *positive definite*, i.e., $ d(\pmb{x,y}) \geq 0 $ for all $ \pmb{x,y} \in V $ and $ d(\pmb{x,y} = 0 \Leftrightarrow \pmb{x} = \pmb{y} $**
    * $ d $ **is symmetric, i.e., $ d(\pmb{x, y})= d(\pmb{y, x}) $ for all $ \pmb{x,y} \in V$**
    * ***Triangle inequality*: $ d(\pmb{x, z}) \leq d(\pmb{x,y}) + d(\pmb{y,z} ) $ for all $\pmb{x,y,z} \in V$**
    * **Very similar $ \pmb{x} $ and $ \pmb{y} $ will result in a large value for the inner product and a small value for the metrix**

## 3.4 Angles and Orthogonality
* **We use the Cauchy-Schawarz inequality to define angles $ \omega $ in inner product spaces between two vectors $ \pmb{x,y} $.**
    * **Assume that $ \pmb{x} \neq \pmb{0}, \pmb{y} \neq \pmb{0}. $ Then
    $$ -1 \leq \langle \frac{\pmb{x,y} \rangle}{\|\pmb{x}\|\|\pmb{y}\|} \leq 1$$ 
Therefore, there exists a unique $ \omega \in [0, \pi], $ with 
$$ \cos \omega = \frac{\langle \pmb{x,y}\rangle}{\|\pmb{x}\pmb{y}\|} $$
The number $ \omega $ is the *angle* between the vectors $ \pmb{x} $ and $ \pmb{y}$.**
* **Orthogonality. Two vectors $ \pmb{x} $ and $ \pmb{y} $ are *orthogonal* if and only if $\langle \pmb{x,y} \rangle = 0$, and we write $ \pmb{x} \bot \pmb{y} $. If additionally $ \| \pmb{x} \| = 1 = \|\pmb{y} \|,$ i.e., the vectors are unit vectors, then $ \pmb{x} $ and $ \pmb{y} $ are *orthonormal***
    * **Vectors that are orthogonal with respect to one inner product do not have to be orthogonal with respect to a different inner product**
* **Orthogonal Matrix. A square matrix $\pmb{A} \in \mathbb{R}^{n \times n} $ is an *orthogonal matrix* if and only if its columns are orthonormal so that
$$ \pmb{A}\pmb{A}^{T}= \pmb{I} = \pmb{A}^{T} \pmb{A} $$
which implies that 
$$ \pmb{A}^{-1} = \pmb{A}^{T} $$
i.e., the inverse is obtained by simply transposing the matrix**
    * **Orthogonal matrices $ \pmb{A} $ with $ \pmb{A}^{T}= \pmb{A}^{-1} $ preserves both angles and distances since 
$$ \| \pmb{Ax} \|^{2}= (\pmb{Ax}^{T}) (\pmb{Ax}) = \pmb{x}^{T} \pmb{A}^{T} \pmb{A} \pmb{x} = \pmb{x}^{T} \pmb{A}^{T} \pmb{A} \pmb{x} = \pmb{x}^T \pmb{I} \pmb{x} = \pmb{x}^{T} \pmb{x} = \|\pmb{x}\|^{2} $$
preserves distances and,
$$ \cos \omega = \frac{(\pmb{Ax})^{T}(\pmb{Ay})}{\|\pmb{Ax}\|\|\pmb{Ay}\|} =\frac{\pmb{x}^{T} \pmb{A}^{T} \pmb{A} \pmb{y}}{\sqrt{\pmb{x}^{T} \pmb{A}^{T} \pmb{Ax} \pmb{y}^{T} \pmb{A}^{T} \pmb{Ay}}} = \frac{\pmb{x}^{T}\pmb{y}}{\|\pmb{x}\|\|\pmb{y}\|}$$
which preserves ttge angles**

## 3.5 Orthonormal Basis
* **Orthonormal Basis. Consider an *n*-dimensional vector space $V$ and a basis $ \{\pmb{b}_1, \dots, \pmb{b}_n\} $ of $V$. If
$$ \langle \pmb{b}_i, \pmb{b}_j \rangle = 0 for i \neq j $$
$$ \langle \pmb{b}_i, \pmb{b}_i \rangle = 1$$
for all $ i, j =1, \dots , n $ then the basis is called an *orthonormal basis*(ONB). If on ly the first criteria is satisfied, then the basis is called an *orthogonal basis*. Note that the second criteria implies that every vector has length/norm 1.**


## 3.6 Orthogonal Complement
* **Consider a D-dimensional vector space $V$ and an $M$ dimensional subspace $ U \subseteq V$. Then its *orthogonal complement* $ U^{\bot} $ is a $ (D-M) $ -dimensional subspace of $V$ and contains all vectors in $V$ that are orthogonal to every vector in $U$.**
* **Furthermore, $ U \cap U^{\bot} = \{\pmb{0}\} $ so that any vector $ \pmb{x} \in V $ can be uniquely decomposed into 
$$ \pmb{x} = \sum_{m=1}^{M} \lambda_m \pmb{b}_m + \sum_{j=1}^{D-M} \psi_j \pmb{b}_j^{\bot}, \lambda_m, \psi_j \in \mathbb{R} $$
where $ (\pmb{b}_1, \dots, \pmb{b}_M)$ is a basis of $U$ and $(\pmb{b}_1^{\bot},\dots, \pmb{b}_{D-M}^{\bot}) $ is a basis of $ U^{\bot} $.**


## 3.7 Inner Product of Functions
* **An inner product of two functions $ \mathcal{u}: \mathbb{R} \to \mathbb{R} $ and $ v: \mathbb{R} \to \mathbb{R} $ can be defined as the definite integral
$$ \langle \mathcal{u,v} \rangle := \int_a^b \mathcal{u(x)v(x)dx}$$
for lower and upper limits $a, b < \infty $**
* **If the equation evaluates to $0$, the functions $ \mathcal{u} $ and $ \mathcal{v} $ are orthogonal.**

## 3.8 Orthogonal Projections
* **Projection.Let $V$ be a vector space and $ U \subseteq V $ a subspace of $V$. A linear mapping $\pi : V \to U $ is called a *projection* if $ \pi^{2} = \pi \circ \pi = \pi $**

### 3.8.1 Projection onto One-Dimensional Subspaces(Lines)
* **Assume we are given a line(one-dimensional subspace) through the origin with basis vector $ \pmb{b} \in \mathbb{R}^{n} $. The line is a one-dimensional subspace $ U \subseteq \mathbb{R}^{n} $ spanned by $\pmb{b}$. When we project $ \pmb{x} \in \mathbb{R}^{n} $ onto $U$, we seek the vector $ \pi_U(\pmb{x}) \in U $ that is closest to $\pmb{x}$**

    * **The projection $ \pi_U(\pmb{x}) $ is closest to $\pmb{x}$, where "closest" implies that the distance $ \|\pmb{x}-\pi_U(\pmb{x})\| $ is minimal/**
    * **The segement $ \pi_U(\pmb{x}) - \pmb{x} $ from $\pi_U(\pmb{x}) $ to $\pmb{x} $ os orthogonal to $U$, and therefore the basis vector $\pmb{b}$ of $U$/, which yields
$$ \langle \pi_U(\pmb{x})-\pmb{x}, \pmb{b} \rangle = 0$$**
    * **The projection $ \pi_U(\pmb{x}) $ of $\pmb{x}$ onto $U$ must be an element of $U$ and, therefore, a multiple of the basis vector $\pmb{b}$ that spaces $U$. Hence, $\pi_U(\pmb{x})=\lambda \pmb{b} $ for some $\lambda \in \mathbb{R}$**
    * **Steps to find the coordinates and the projection matrix**
        1. **Find the coordiante $\lambda$. Since orthogonality condition yields that
$$ \langle \pmb{x} - \pi_U(\pmb{x}), \pmb{b} \rangle = 0 \overset{\pi_U(\pmb{x})=\lambda \pmb{b}}{\Leftrightarrow} \langle \pmb{x} - \lambda \pmb{b}, \pmb{b} \rangle = 0 $$
we can now explot the bilineariy of the inner product and arrive at
$$ \langle \pmb{x}, \pmb{b} \rangle - \lambda \langle \pmb{b}, \pmb{b} \rangle = 0 \Leftrightarrow \lambda = \frac{\langle \pmb{x,b} \rangle}{\langle \pmb{b,b} \rangle}=\frac{\langle\pmb{b,x}\rangle}{\|\pmb{b}\|^2}$$ 
If we choose $ \langle \cdot, \cdot \rangle$ to be the dot product, we obtain 
$$ \lambda = \frac{{b}^{T}\pmb{x}}{\pmb{b}^{T}\pmb{b}} = \frac{\pmb{b}^{T}\pmb{x}}{\|\pmb{b}\|^2} $$
If $\|b\|=1$, then the coordinate $\lambda$ of the projection is given by $\pmb{b}^{T}\pmb{x}$**
        1. **Finding the projection point $ \pi_U(\pmb{x}) \in  U $. Since $\pi_U(\pmb{x})= \lambda \pmb{b}$, we immediately obtain that
$$ \pi_U(\pmb{x})=\lambda \pmb{b} = \frac{\langle \pmb{x, b} \rangle}{\|\pmb{b}\|^2}\pmb{b}=\frac{\pmb{b}^{T} \pmb{x}}{\|\pmb{x}\|^2}\pmb{b}$$
where the last equality holds fot the dot product only. We can also compute the lenght of $ \pi_U(\pmb{x}) $ as
$$ \| \pi_U(\pmb{x})\| = \|\lambda \pmb{b} \| = \mid \lambda \mid \|\pmb{b}\| $$
This shows that $ \lambda $ is the coordinate of $ \pi_u(\pmb{x}) $ with respect to the basis vector $ \pmb{b} $ that spans our one-dimensional subspace $U$. If we use the dot product as an inner product, we get
$$ \|\pi_U(\pmb{x})\| = \frac{\mid \pmb{b}^{\bot} \pmb{x}\mid}{\|\pmb{b}\|^2}\|\pmb{b}\|=\mid \cos \omega \mid \| \pmb{x}\| \|\pmb{b}\|\frac{\|\pmb{b}\|}{\|\pmb{b}\|^2}=\mid\cos\omega\mid\|\pmb{x}\|$$
$\omega$ is the angle between $\pmb{x}$ and $\pmb{b}$.**

        1. **Finding the projection matrix $ \pmb{P}_pi $. There exists a projection matrix $\pmb{P}_\pi$, such that $\pi_U(\pmb{x}) = \pmb{P}_{\pi}{\pmb{x}} $ with the dot product as inner product and 
        $$ \pi_U(\pmb{x}) = \lambda \pmb{b} = \pmb{b} \lambda= \pmb{b} \frac{\pmb{b}^{T}\pmb{x}}{\|\pmb{b}\|^{2}}=\frac{\pmb{b}\pmb{b}^{T}}{\|\pmb{b}\|^2}\pmb{x} $$
$$ \pmb{P}_pi = \frac{\pmb{b}\pmb{b}^{T}}{\|\pmb{b}\|^2} $$
Note that $ \pmb{b} \pmb{b}^{T} $ is a symmetric matrix(of rank 1), and $\|b\|^2 = \langle \pmb{b},\pmb{b} \rangle $ is a scalar**

### 3.8.2 Projection onto General Subspaces

* **Assume that $ (\pmb{b}_1,\dots,\pmb{b}_m) $ is an ordered basis of $U$. Any projection $\pi_U(\pmb{x}) $ onto $U$ is necessarily an element of $U$. They can be represented as linear combination of the basis vectors $\pmb{b}_1,\dots,\pmb{b}_m $ of $U$, such that $\pi_U(\pmb{x})=\sum_{i=1}^{m}\lambda_i\pmb{b}_i $**
* **Find the projection $ \pi_U(\pmb{x}) $ and the projection matrix $\pmb{P}_{\pi}$:**
    1. **Find the coordinates $\lambda_1,\dots,\lambda_m $ of the projection(with respect to the basis of $U$, such that the linear combination
$$ \pi_U(\pmb{x}) = \sum_{i=1}^{m} \lambda_i \pmb{b}_i = \pmb{B}\pmb{\lambda}$$
$$  \pmb{B} = [\pmb{b}_1,\dots,\pmb{b}_m]\in \mathbb{R}^{n \times m}, \pmb{\lambda}=[\lambda_1,\dots,\lambda_m]\in \mathbb{R}^{m}$$
is closest to $\pmb{x}\in \mathbb{R}^{n} $. "Closest" means "minimum distance", which implies that the vector connecting $\pi_U(\pmb{x}) \in U $ and $ \pmb{x} \in \mathbb{R}^{n} $ must be orthogonal to all basis vectors of U. We obtain a homogeneous linear equation system
$$
\begin{bmatrix}
\pmb{b}_1^T\\
\vdots\\
\pmb{b}_m^T\\
\end {bmatrix}
\begin {bmatrix}
\pmb{x-B\lambda}\\
\end{bmatrix}
=
0
\Leftrightarrow
\pmb{B}^{T}(\pmb{x}-\pmb{B\lambda})=\pmb{0}\\
\Leftrightarrow
\pmb{B}^{T}\pmb{B}\pmb{\lambda}=\pmb{B}^{T}\pmb{x}
$$
We can solve for the coefficients/coordinates using
$$ \pmb{\lambda}=(\pmb{B}^{T}\pmb{B})^{-1}\pmb{B}^{T}\pmb{x}$$. the matrix $ (\pmb{B}^{T}\pmb{B})^{-1}\pmb{B}^{T} $ is also called the *pseudo-inverse***
    1. **Find the projection $\pi_U(\pmb{x})\in U $.**
    $$ \pi_U(\pmb{x})=\pmb{B}(\pmb{B}^{T}\pmb{B})^{-1}\pmb{B}^{T}\pmb{x}$$
    1. **Find the projectopn $\pmb{P}_{\pi}$**
$$ \pmb{P}_pi=\pmb{B}(\pmb{B}^{T}\pmb{B})^{-1}\pmb{B}^{T}$$
* ***least squares solution*: We can find approximate solution. The idea is to find the vector in the subspace spanned by the columns of $\pmb{A}$ that is closest to $\pmb{b}$, i.e., we compute the orthogonal projection of $ \pmb{b} $ onto the subpace spanned by the columns of $\pmb{A}$**

### 3.8.3 Gram-Schmidt Orthogonalization
* **The *Gram-Schmidt orthogonalization* method iteratively constructs an orthogonal basis $(\pmb{u}_1,\dots,\pmb{u}_n)$ of $V$ as follows"**
    $$ \pmb{u}_1:= \pmb{b}_1;$$
    $$ \pmb{u}_k:= \pmb{b}_k-\pi_{span[\pmb{u}_1,\dots,\pmb{u}_{k-1}]}(\pmb{b}_k), k = 2, \dots, n $$
    1. **The $kth$ basis vector $\pmb{b}_k $ is projected onto the subspace spanned by the first *k*-1 constructed orthogonal vectors $ \pmb{u}_1,\dots,\pmb{u}_{k-1} $**
    1. **This projection is then subtracted from $\pmb{b}_k $ and yields a vector $\pmb{u}_k $ that is orthogonal to the $(k - 1) $-dimensional subspaced spanned by $\pmb{u}_1,\dots, \pmb{u}_{k-1}$.**
    1. **Repeating thid procedure fot all *n* basis vectors $ \pmb{b}_1, \dots, \pmb{b}_n $ yields an orthogonal basis $(\pmb{u}_1, \dots, \pmb{u}_n) $ of $V$. If we normalize $\pmb{u}_k$, we obtain an ONB where $ \|\pmb{u}_k \|=1 $ for $k = 1, \dots, n$**


### 3.8.4  Projection onto Affine Subspaces

* **We subtract the support point $ \pmb{x}_0 $ from $ x $ and from $L$, so that $ \pmb{L} - \pmb{x}_0 = U $ is exactly the vector subspace $U$.** 
* **We can now use the orthogonal projections onto a subspace and obtain the projection $ \pi_U(\pmb{x}-\pmb{x}_0)$**
* **The projection can now be translated back into $ L$ by adding $\pmb{x}_0$, such that we obtain the orthogonal projection onto an affine space $L $ as 
$$ \pi_L(\pmb{x})= \pmb{x}_0 + \pi_U(\pmb{x}-\pmb{x}_0) $$
where $ \pi_U(\cdot) $ is the orthogonal projection onto the subspace $U$, i.e., the direction space of $L$.**

## 3.9 Rotations
* **A *rotation* is a linear ,apping that rotates a plane by an angle of $\theta$ about the origin, i.e., the origin is a fixed point**

### 3.9.1 Rotations in $\mathbb{R}^2$
* **The rotation matrix that performs the basis change into the rotated coordinates $ \pmb{R}(\theta)$ is given as 
$$
\pmb{R}(\theta)
=
\begin {bmatrix}
\Phi(\pmb{e}_1)\space \space \space \Phi(\pmb{e}_2)\\
\end {bmatrix}
= 
\begin {bmatrix}
\cos\theta \space \space \space -\sin\theta\\
\sin\theta \space\space\space \cos\theta\\
\end {bmatrix}
$$**


### 3.9.2 Rotations in $\mathbb{R}^3$
* **In $ \mathbb{R}^3$, we can rotate any two-dimensional plane about a one-dimensional axis**
* **We use the convention that a "counterclockwise"(planar) ratation about an axis refers to a rotation about an axis when we look at the axis "head on", from the end towards the "origin".In $ \mathbb{R}^3$,tgere are therefore three (planar) rotations about the three standard basis vectors**
* **Rotation about the $e_1$ axis
$$ \pmb{R}_1(\theta)=[\Phi(\pmb{e}_1)\space \space \space \Phi(\pmb{e}_2) \space \space \space \Phi(\pmb{e}_3)] 
=
\begin {bmatrix}
1 \space \space \space \space \space \space \space \space \space 0 \space \space \space \space \space \space \space \space \space 0\\
0 \space \space \space \cos\theta \space \space \space -\sin\theta\\
0 \space \space \space \sin\theta \space \space \space \cos\theta\\
\end {bmatrix}**
$$
* **Rotation about the $e_2$ axis
$$
\pmb{R}_2(\theta)
=
\begin {bmatrix}
\cos\theta \quad 0 \quad \sin\theta\\
0 \qquad 1 \qquad 0\\
-\sin\theta \quad 0\quad \cos\theta\\
\end {bmatrix}
$$
If we rotate the $\pmb{e}_1\pmb{e}_3$ plane about the $\pmb{e}_2$ axis, we need to look at the $\pmb{e}_2$ axis from its tip toward the origin**
* **Rotation about the $\pmb{e}_3 $ axis
$$
\pmb{R}_3(\theta)
=
\begin {bmatrix}
\cos\theta \quad -\sin\theta \quad 0\\
\sin\theta \quad \cos\theta \quad 0\\
0 \quad 0 \quad 1\\
\end {bmatrix}
$$


### 3.9.3 Rotations in $ n$ Dimensions
* **Definition: Let $V$ be an $n-$ dimensional Euclidean vector space and $\Phi: V\to V $ an automorphism with transformatio matrix

for $ 1\leq i \leq j \leq n $ and $ \theta \in \mathbb{R}$. Then $\pmb{R}_{ij}(\theta)$ is called a *Givens rotation*. Essentially, $ \pmb{R}_{ij}(\theta) $ is the identity matrix $\pmb{I}_n$ with**
$$ r_{ii} = \cos \theta, \quad r_{ij} = -\sin\theta, \quad r_{ji}=\sin\theta, \quad r_jj=\cos\theta$$
